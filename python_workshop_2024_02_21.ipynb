### Evaluate records to find those with overlapping coverage dates

Select records where both `coverage_start` and `coverage_end` exists - are not missing. Also select records with 'Pittsburgh' in the `municipality` column. Finally, select only relevant columns to better view the data.

Sort the values by `coverage_start`.

```python
pgh_df = df.loc[~df['coverage_start'].isna() & ~df['coverage_end'].isna() & (df['municipality']=='Pittsburgh'), ['name','agency_described','coverage_start','coverage_end']].\
  sort_values('coverage_start').\
  copy()
```

```python
pgh_df
```

Add a column for todays date - experimenting with the idea that we can determine overlap by finding the difference in days between today's date and the coverage dates.

```python
pgh_df['today'] = datetime.date.today()
```

```python
pgh_df
```

```python
pgh_df['coverage_start'] = pd.to_datetime(pgh_df['coverage_start'])
```

```python
pgh_df['coverage_end'] = pd.to_datetime(pgh_df['coverage_end'])
```

```python
pgh_df['today'] = pd.to_datetime(pgh_df['today'])
```

```python
pgh_df['today-start'] = (pgh_df['today'].dt.date - pgh_df['coverage_start'].dt.date).dt.days
```

```python
pgh_df['today-end'] = (pgh_df['today'].dt.date - pgh_df['coverage_end'].dt.date).dt.days
```

Implement a hack to accomplish the task:

Using the **filtered, sorted** dataset `pgh_df`...

1.   Create a column `idx` that numbers the rows in ascending order starting with 0.
2.   Create a copy of the data frame (drop the new date and days columns created above) and name the new data frame `pgh_actual` to indicate that these coverage dates match the `name` and `agency_described` columns.
3.   Make a second copy of the data frame, called `pgh_previous` and 1) rename `coverage_start` and `coverage_end` to `previous_start` and `previous_end`, respectively 2) drop `name` and `agency_described`.
4. Overwrite the `idx` column by adding 1 to the existing `idx` value.
5. Merge `pgh_actual` with `pgh_previous` on `idx` using a left merge with `pgh_actual` listed first. This means all records from `pgh_actual` will be preserved and all matching records from `pgh_previous` will be merged with it so that the each record will have its `coverage_start` and `coverage_end` values as well as the previous records dates as `previous_start` and `previous_end`.
6.   Rearrange the columns to be in logical order. Select relevant columns excluding `idx`.
7.   Finally, check if the `previous_end` is greater than or equal to the `coverage_start`. Create a new variable `overlap` using a list comprehension to flag records with overlapping dates.
