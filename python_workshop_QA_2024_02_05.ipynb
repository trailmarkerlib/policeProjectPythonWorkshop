### Read marijuana csv file

```python
drive_url = '/content/drive/MyDrive/_CAASI/Grief to Action - PDAP/'
```

```python

df = pd.read_csv(drive_url + 'PDAP Data Sources.csv')
```

```python
df.info()
```

```python
df.record_type.unique()
```

```python
use_of_force = df.loc[df['record_type'].isin(['Use of Force Reports', 'Officer Involved Shootings'])].\
                  reset_index().\
                  copy()
```

```python
use_of_force.info()
```

```python
munic_agg = use_of_force.groupby(['state', 'municipality']).\
              aggregate(rec_count=('record_type', 'count')).\
              reset_index().\
              copy()
```

```python
munic_agg
```

#### Agency Type

```python
df.agency_type.unique()
```

```python
agency_df = df.loc[df['agency_type'].isin(['law enforcement/police', 'corrections', 'court'])].copy()
```

```python
agency_agg = agency_df.loc[:, ['state', 'municipality', 'agency_type']].\
              drop_duplicates().\
              groupby(['state', 'municipality']).\
              aggregate(rec_count=('agency_type', 'count')).\
              reset_index().\
              copy()
```

```python
agency_agg
```

```python
agency_agg.loc[agency_agg['rec_count'] > 2]
```
